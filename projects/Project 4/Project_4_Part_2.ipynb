```python
states=[]

for i in jobs.job_addr:
    if 'CA' in i:
        states.append('CA')
    elif 'NSW' in i:
        states.append('NSW')
    elif 'VIC' in i:
        states.append('VIC')
    else:
        states.append('i')
        
jobs['state'] = states

jobs.state.unique()
```

```python
jobs.salary_bins_int.where?
```

```python
#How would I handle this if I had three currencies?
#Why is it behaving this way?

jobs['adj_salary_bin'] = jobs.salary_bins_int.where(jobs.state == 'CA', jobs.salary_bins_int*0.77)
jobs
```
