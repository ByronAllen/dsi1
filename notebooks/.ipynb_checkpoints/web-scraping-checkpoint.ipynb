```python
html_req = requests.get ('http://www.bom.gov.au/nsw/forecasts/sydney.shtml')
soup = BeautifulSoup(html_req.text, 'lxml')
```

```python
weather_dict = {} 
```

```python
syd_weather = soup.find('div', class_='day main')

first_day = syd_weather.find('h2').string
max_temp_tdy = syd_weather.find('em', class_="max").string

weather_dict[first_day] = max_temp_tdy
```

```python
syd_weather_rem = soup.find_all('div', class_='day')
for i in syd_weather_rem:
    days = i.find('h2').string
    weather = i.find('em', class_="max").string
    weather_dict[days] = weather
```

```python
weather_dict
```

### Example from local file:

```python
import json
```

```python
decoder = json.JSONDecoder()
yelp_data_as_list = open('yelp.json').readlines()
```

```python
json.loads(yelp_data_as_list[0])
```

```python
stars = []
mentions_veggies = []

for review in yelp_data_as_list:
    parsed = json.loads(review)
    stars.append(parsed['stars'])
    mentions_veggies.append('vegetable' in parsed['text'].lower())

zip(stars, mentions_veggies)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(
    data={
        'Stars': stars,
        'Mentions_Veggies': mentions_veggies
    }
)
```

```python
import seaborn
```

```python
seaborn.distplot(df[df.Mentions_Veggies].Stars)
seaborn.distplot(df[df.Mentions_Veggies==False].Stars)
```
