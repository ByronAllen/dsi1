```python
def time_convert(x):
    
    duration = []
    
    for t in x:
        
        if ':' not in t:
            fin = int(t)*60
        
        else:    
            spliting = t.split(':')
        
            if spliting[0] == '':
                min_to_sec = 0
            else:
                min_to_sec = int(spliting[0])*60 
                
            if spliting[1] == '':
                seconds = 0 
            else:
                seconds = int(spliting[1])
        
            fin = min_to_sec + seconds
        
        duration.append(fin)
        
    return duration
```

```python
time_convert(bb.time)
```

```python
bb['time'] = time_convert(bb.time)
```
