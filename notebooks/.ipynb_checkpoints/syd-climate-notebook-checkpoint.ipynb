```python
plt.style.use('classic')

with plt.style.context('fivethirtyeight'):
    plt.plot(pd.rolling_mean(sycl.MANGROVE_MOUNTAIN, 5))
    plt.plot(pd.rolling_mean(sycl.CAMDEN_AIRPORT, 5))
    plt.plot(pd.rolling_mean(sycl.BELLAMBI, 5))
    plt.legend()
    
plt.title("Sydney Climate")
plt.xlabel("Date")
plt.ylabel("Temperature ($^\circ$C)")


ax2 = plt.twinx()
ax2.set_ylabel("F")

#t = None
#s2 = None
#ax2.plot(t, s2, 'r.')

plt.show()
```
