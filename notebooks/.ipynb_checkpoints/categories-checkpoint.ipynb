# Categories

```python
%matplotlib inline
import pandas as pd
import sklearn.linear_model
```

```python
sac = pd.read_csv('Sacramentorealestatetransactions.csv')
```

Adding a column that is a category derived from column that is a string

```python
sac['dwelling_type'] = sac.type.astype('category')
```

Convert a category array into several columns that list binary outcome

```python
sac_w_dummies = pd.get_dummies(sac, columns=['dwelling_type'])
sac_w_dummies
```

If we attempt a linear regression using all 4 dummy variables, there are an infinite number of options for the zero-intercept point. 

```python
ols = sklearn.linear_model.LinearRegression()
columns = ['dwelling_type_Condo', 'dwelling_type_Multi-Family', 'dwelling_type_Residential', 'dwelling_type_Unkown']
ols.fit(sac_w_dummies[columns],sac_w_dummies.price)
zip(columns,ols.coef_)
```

Sklearn is able (probably has) to choose the zero-intercept to be the mean price of the properties which are all zero in the columns.......

```python
ols.intercept_
```

```python
sac_w_dummies[sac_w_dummies.dwelling_type_Unkown == 1]
```

```python
sac_w_dummies[sac_w_dummies.dwelling_type_Condo == 1].price.mean()
```

## Other options on get_dummies

```python
pd.get_dummies(sac, columns=['zip'])
```

This will create a table with far too many columns... 68 columns in fact. This also occupies a lot of memory.

```python
sac.zip.nunique()
```

```python
sac_w_postcodes = pd.get_dummies(sac,
                                 columns=['zip'],
                                 sparse=True)
```

```python
sac_w_postcodes.zip_95603.mean()
```

```python
columns_to_regress_on = [x for x in sac_w_postcodes.columns if x.startswith('zip_') and x != 'zip_95603']
columns_to_regress_on
```

```python
ols.fit(sac_w_postcodes[columns_to_regress_on], sac_w_postcodes.price)
zip(columns_to_regress_on, ols.coef_)
```
