# The following three ways of definging not_func are equal, yet not as concise.

```python
def not_func(x):
    return not x

def not_func(x): return not x

not_func = lambda x: not x

cheapskate = mill.map(not_func)
```

```python
low_value_prod = sales_data[cheapskate]
```

```python
matplotlib.pyplot.scatter(low_value_prod.Q1_sales_total,
                          low_value_prod.growth_rate_log)
```

```python
sum(low_value_prod.Q1_sales_total)
```

```python
low_value_prod.Q1_sales_total.sum()
```

```python
sales_data[mill].Q1_sales_total.sum()
```

```python
labels = ['High Value Products', 'Low Value Products']
sizes = [sales_data[mill].Q1_sales_total.sum(),
        low_value_prod.Q1_sales_total.sum()]
colors = ['gold', 'lightskyblue']
explode = (0.2, 0)  

matplotlib.pyplot.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
```

# Growth rate analysis

```python
sales_data.growth_rate_log.mean()
```

```python
five_bin_cut = pandas.cut(sales_data.growth_rate_log, 5)
```

```python
hist_data = pandas.Series(data=sorted(five_bin_cut.value_counts()))
matplotlib.pyplot.plot(hist_data.values)
```

```python
n, bins, patches = matplotlib.pyplot.hist(sales_data.growth_rate_log, 30, normed=1, facecolor='green', alpha=0.75)
```
