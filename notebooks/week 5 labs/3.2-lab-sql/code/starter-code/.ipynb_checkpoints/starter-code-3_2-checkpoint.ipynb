## 2: Products

What products is this company selling? The `products` and `categories` tables contain information to answer this question.

Use a combination of SQL queries and Pandas merge to answer the following questions:

- What categories of products is the company selling?
- How many products per category does the catalog contain?
- Let's focus only on products that have not been discontinued => how many products per category?
- What are the most expensive 5 products (not discontinued)?
- How many units of each of these 5 products are there in stock?
- Draw a pie chart of the categories, with slices that have the size of the number of products in that category (use non discontinued products)

```python
%%sql postgresql://dsi_student:gastudents@dsi.c20gkj5cvu3l.us-east-1.rds.amazonaws.com/northwind
select * from INFORMATION_SCHEMA.COLUMNS where table_name = 'categories';
```

### 2.a: What categories of products is the company selling?

Remember that PostgreSQL is case sensitive.

```python
pd.DataFrame?
```

```python
categories = %sql select "CategoryID", "CategoryName", "Description" from categories;
categories = pd.DataFrame(data=categories, columns=["CategoryID", "CategoryName", "Description"])
categories
```

### 2.b: How many products per category does the catalog contain?

Keep in mind that you can cast a %sql result to a pandas dataframe using the `.DataFrame()` method.

```python
%%sql postgresql://dsi_student:gastudents@dsi.c20gkj5cvu3l.us-east-1.rds.amazonaws.com/northwind
select distinct("ProductName") from products;
select count("ProductName") from products;
```

```python
prod_per_cat = %sql select count("ProductName"), "CategoryID" from products group by "CategoryID";
prod_per_cat = pd.DataFrame(data=prod_per_cat, columns= ['prod_per_cat','CategoryID'])
prod_per_cat
```

```python
categories.merge(prod_per_cat)
```

### 2.c: How many not discontinued products per category?

```python
%%sql postgresql://dsi_student:gastudents@dsi.c20gkj5cvu3l.us-east-1.rds.amazonaws.com/northwind
select count(*) from products where "Discontinued" = 1;
```

### 2.d: What are the most expensive 5 products (not discontinued)?

```python
%%sql postgresql://dsi_student:gastudents@dsi.c20gkj5cvu3l.us-east-1.rds.amazonaws.com/northwind
select * from products where "Discontinued" = 1 order by "UnitPrice" desc limit 5;
```

### 2.e: How many units of each of these 5 products are there in stock?

```python
%%sql postgresql://dsi_student:gastudents@dsi.c20gkj5cvu3l.us-east-1.rds.amazonaws.com/northwind
select "ProductName", "UnitsInStock" from products where "Discontinued" = 1 order by "UnitPrice" desc limit 5;
```

### 2.f: Pie Chart

Use pandas to make a pie chart plot.

```python
in_stock = %sql select "ProductName", "UnitsInStock" from products where "Discontinued" = 1 order by "UnitPrice" desc;
in_stock = pd.DataFrame(data=in_stock,columns=["ProductName", "UnitsInStock"])
in_stock = in_stock[in_stock.UnitsInStock !=0]
in_stock
```

```python
product_names_aslist = list(in_stock.ProductName)
```

```python
plt.pie(in_stock.UnitsInStock, labels=product_names_aslist)
plt.show()
```

### 3.b: How many orders per year?

```python
%%sql postgresql://dsi_student:gastudents@dsi.c20gkj5cvu3l.us-east-1.rds.amazonaws.com/northwind
select "column_name" from INFORMATION_SCHEMA.COLUMNS where table_name = 'orders';
```

```python
orders = %sql select * from orders;
col = %sql select "column_name" from INFORMATION_SCHEMA.COLUMNS where table_name = 'orders';
col = [x[0] for x in col]
orders = pd.DataFrame(data=orders,columns=col)
orders.info()
```

```python
#s = '1996-07-08'
#s.split('-')[0]

orders['year'] = [x.split('-')[0] for x in orders.OrderDate.astype(str)]
orders['month'] = [x.split('-')[1] for x in orders.OrderDate.astype(str)]
dictionary = {'01':1,'02':1,'03':1,'04':2,'05':2,'06':2,'07':3,'08':3,'09':3,'10':4,'11':4,'12':4}
orders['quarter'] = orders.month.map(dictionary)
```

```python
orders.pivot_table(index='year', values='OrderID', aggfunc='count')
```

### 3.c: How many orders per quarter?

Make a line plot for these.

```python
orders.pivot_table(index='quarter', values='OrderID', aggfunc='count')
```
