## 2. Data Exploration

Use the 3 dataframes to answer the following questions:

1. How many employees are there in the company?
- How many messages are there in the database?
- Convert the timestamp column in the messages. When was the oldest message sent? And the newest?
- Some messages are sent to more than one recipient. Group the messages by message_id and count the number of recepients. Then look at the distribution of recepient numbers.
    - How many messages have only one recepient?
    - How many messages have >= 5 recepients?
    - What's the highest number of recepients?
    - Who sent the message with the highest number of recepients?
- Plot the distribution of recepient numbers using Bokeh.

```python
#number of employees in compnay
results = c.execute("select count(eid) from EmployeeBase")
results.fetchall()
```

```python
#number of messages are there in the database MessageBase
results = c.execute("select count(*) from MessageBase")
results.fetchall()
```

```python
#Convert the timestamp column in the messages. When was the oldest message sent? 
import datetime 
results = c.execute("select min(unix_time) from MessageBase")
datetime.datetime.fromtimestamp(results.fetchall()[0][0]).strftime('%Y-%m-%d %H:%M:%S')
```

```python
#And the newest?\
results = c.execute("select max(unix_time) from MessageBase")
datetime.datetime.fromtimestamp(results.fetchall()[0][0]).strftime('%Y-%m-%d %H:%M:%S')
```

```python
#Some messages are sent to more than one recipient. Group the messages by message_id and count the number of recepients. 
#Then look at the distribution of recepient numbers.
```

```python
#How many messages have only one recepient?
results = c.execute("select mid, count(to_eid) from RecipientBase group by mid")
messages = results.fetchall()
total=0

for i in messages:
    if i[1] == 1:
        total += 1
        
print total
```

```python
#How many messages have >= 5 recepients?

results = c.execute("select mid, count(to_eid) from RecipientBase group by mid")
messages = results.fetchall()
total=0

for i in messages:
    if i[1] >= 5:
        total += 1
        
print total
```

```python
#What's the highest number of recepients?
#How many messages have >= 5 recepients?

results = c.execute("select mid, count(to_eid) from RecipientBase group by mid")
messages = results.fetchall()

unique_rec_per_mid = []

for i in messages:
    unique_rec_per_mid.append(i[1])
        
max(unique_rec_per_mid)
```

```python
#Who sent the message with the highest number of recepients?
results = c.execute("select mid, count(to_eid) from RecipientBase group by mid having count(to_eid) = 57")
results.fetchall()
```

```python
results = c.execute('select from_eid from MessageBase where mid = 12116')
eid = results.fetchall()[0][0]
results = c.execute("select name from EmployeeBase where eid = %s" % eid)
print results.fetchall()[0][0]

results = c.execute('select from_eid from MessageBase where mid = 12151')
eid = results.fetchall()[0][0]
results = c.execute("select name from EmployeeBase where eid = %s" % eid)
print results.fetchall()[0][0]
```
