## 3.  Perform a K-Means Clustering 

#### 3.1 Split the dataset into the attribute data and class labels

```python
list(airports.columns)
```

```python
attributes = airports[['Year',
                     'Departure Cancellations',
                     'Arrival Cancellations',
                     'Departure Diversions',
                     'Arrival Diversions']]

classes = airports.Airport
```

#### 3.2 Conduct the k-means clustering & 3.3 Compute the labels and centroids

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
k=3
kmeans = KMeans(n_clusters=k)
kmeans.fit(attributes)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
silhouette_score(airports, labels, metric='euclidean')
```

```python
airports['Cluster_K3'] = labels
airports.head()
```

#### 3.4 Format the resulting predicted "y" variables and check the accuracy score, classification report, and confusion matrix

```python
zip(kmeans.fit_predict(attributes), labels)
```

## 4. Extra Trees Classifier

```python
from sklearn.cross_validation import cross_val_score
```

```python
X = attributes
y_airports = classes
y_cluster_k3 = airports.Cluster_K3
```

```python
etc = ExtraTreesClassifier(random_state=42)

print 'Scoring with y_airports:'
score = cross_val_score(etc, X, y_airports, scoring='accuracy', cv=5)
print score.mean()*1
print score.std()
print 

print 'Scoring with y_cluster_k3:'
score = cross_val_score(etc, X, y_cluster_k3, scoring='accuracy', cv=5)
print score.mean()*1
print score.std()
```

```python
etc.feature_importances_
```

## 5. KNN Classifier

```python
knn = KNeighborsClassifier()

print 'Scoring with y_airports:'
score = cross_val_score(knn, X, y_airports, scoring='accuracy', cv=5)
print score.mean()*1
print score.std()
print 

print 'Scoring with y_cluster_k3:'
score = cross_val_score(knn, X, y_cluster_k3, scoring='accuracy', cv=5)
print score.mean()*1
print score.std()
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X,y_airports)
print 'Impact of attributes on y_airports: ' 
lr.coef_
```

```python
lr2 = LogisticRegression()
lr2.fit(X,y_cluster_k3)
print 'Impact of attributes on y_cluster_k3: ' 
lr2.coef_
```

## 6. Hierarchical Clustering

```python
from scipy.cluster.hierarchy import dendrogram, linkage, cophenet, fcluster
from scipy.spatial.distance import pdist
```

#### 6.1 Create the linkage for clustering

```python
X = X.as_matrix(columns=None)
Z = linkage(X, 'ward')
c, coph_dists = cophenet(Z, pdist(X))
c
```

```python
coph_dists.shape
#coph_dists
```

#### 6.2 Plot the Denrogram

```python
plt.figure(figsize=(30, 10))
plt.title('Dendrogram')
plt.xlabel('Index Numbers')
plt.ylabel('Distance')
dendrogram(
    Z,
    leaf_rotation=90.,  
    leaf_font_size=8.,
)
plt.show()
```

```python
plt.title('Truncated Dendrogram')
plt.xlabel('Index Numbers')
plt.ylabel('Distance')
dendrogram(
    Z,
    truncate_mode='lastp',  
    p=15,  
    show_leaf_counts=False,  
    leaf_rotation=90.,
    leaf_font_size=12.,
    show_contracted=True,  
)
plt.show()
```

```python
max_d = 15
clusters = fcluster(Z, max_d, criterion='distance')
clusters
```
