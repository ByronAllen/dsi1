```python
import pandas as pd
import numpy as np

from sklearn.cluster import DBSCAN, KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
%matplotlib inline 
```

## 1. Load the data

```python
from sklearn.datasets import load_iris
X, y = load_iris().data, load_iris().target
print X.shape
print y.shape
```

## 2. Format and graph the data

#### 2.1 Plot the data

```python
import seaborn as sb
sb.pairplot(pd.DataFrame(X))
```

#### 2.2 Select columns of interest and standardize the data

```python
X_standard = StandardScaler().fit_transform(X)
```

## 3. DBSCAN

#### 3.1 Setup the DBSCAN from scikit

```python
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan.fit(X_standard)
```

#### 3.2 Calculate the Labels

```python
labels = dbscan.labels_
print(labels)
```

#### 3.3 Calculate the Core Samples

```python
core_samples = np.zeros_like(labels, dtype = bool) 
core_samples
```

```python
core_samples[dbscan.core_sample_indices_] = True 
```

```python
print(core_samples)
```

```python
set(labels)
```

```python
len(set(labels))
```

```python
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_clusters_
```

```python
from sklearn import metrics

print('Estimated number of clusters: %d' % n_clusters_)

#Homogeneity: If one cluster Homogeneity is = 1
print("Homogeneity: %0.3f" % metrics.homogeneity_score(y, labels))

#Id of outliers 
print("Completeness: %0.3f" % metrics.completeness_score(y, labels))

#v = 2 * (homogeneity * completeness) / (homogeneity + completeness)
print("V-measure: %0.3f" % metrics.v_measure_score(y, labels))
```

#### 3.4 Plot the resulting clusters

Now, repeat the above steps for various levels of minimum points and epsilon and see how that affects the results of the DBSCAN algorithm 

```python
plt.scatter(X_standard[:,0], X_standard[:,1], c=dbscan.labels_, s=50)
```

```python
unique_labels = np.unique(labels)
colors = plt.cm.Spectral(np.linspace(0,1, len(unique_labels)))
```

```python
unique_labels
```

```python
colors
```

```python
core_samples
```

```python
for (label, color) in zip(unique_labels, colors):
    
    class_member_mask = (labels == label)
    n = X_standard[class_member_mask & core_samples]
    
    #Core samples:
    plt.plot(n[:,0],n[:,1], 'o', markerfacecolor = color, 
             markersize = 10)
    
    #Outliers:
    n = X_standard[class_member_mask & ~core_samples]
    plt.plot(n[:,0],n[:,1], 'o', markerfacecolor = color, 
             markersize = 5)
```

#### 3.5 Perform a K-means Clustering for comparison

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
k=3
kmeans = KMeans(n_clusters=k)
kmeans.fit(X_standard)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
print
```

#### 3.6 Plot the resulting clusters

**Hint**: Display the results by cluster color and plot the centroids over the original plot for clarity

```python
print 'silhouette_score: ', silhouette_score(X_standard, labels, metric='euclidean')
plt.scatter(X_standard[:,0], X_standard[:,1], c=labels, s=50)

plt.scatter(centroids[:, 0], centroids[:, 1],
            marker='x', s=169, linewidths=3,
            color='g', zorder=10)
```
