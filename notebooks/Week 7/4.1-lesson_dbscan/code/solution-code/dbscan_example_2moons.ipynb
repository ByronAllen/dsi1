#### Set up and generate data

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN, KMeans

#plt.style.use('ggplot')
%matplotlib inline
```

```python
X, label = make_moons(n_samples=200, noise=0.1, random_state=19)
print X[:5,]
```

#### Train the model

```python
#model = DBSCAN?
```

```python
model = DBSCAN(eps=0.25, min_samples=12).fit(X)
print model
```

#### Access the cluster labels and core points

```python
model.labels_
```

```python
model.core_sample_indices_
```

#### Plot the clusters in feature space

```python
plt.scatter(X[:,0], X[:,1], c=model.labels_, s=50)
```

#### K-means for comparison

```python
kmeans = KMeans(n_clusters = 2).fit(X)
```

```python
labels_km = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
plt.scatter(X[:,0], X[:,1], c = labels_km, s=50)
```
